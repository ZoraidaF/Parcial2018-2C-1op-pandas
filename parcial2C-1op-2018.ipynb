## 1a) indicar los titulos de categoria men para los cuales el importe de venta supera el promedio mensual de venta d los productos de la misma categoria

```python
productos_men=productos[productos.categoria=='men']
productos_men
```

```python
combinado=pd.merge(productos_men,sales,how='inner',on='product_id')
combinado
```

### 2b) indicar el top-10 de productos que se vendieron mayor cantidad de dias de forma consecutiva

```python
b=sales.drop_duplicates(['dia', 'mes','año','product_id'], keep='first')
b
```

```python
b.rename(columns={'año':'year','mes':'month','dia':'day'},inplace=True)
```

```python
b['timestamp']=pd.to_datetime(b[['year', 'month', 'day']])
b
```

```python
b['dia_previo']=b.groupby('product_id').timestamp.shift(1)
b
```

```python
b['dias_consecutivos'] = b.apply(lambda row: int(0) if (row['timestamp']-row['dia_previo'])==pd.Timedelta('1 days') else int(1), axis=1)
```

```python
b
```

```python
b['groups'] = b.groupby('product_id')['dias_consecutivos'].cumsum()
b
```

```python
 #muestro solo aquellos que son consecutivos
consecutivos=b[b['groups'] == 1]
consecutivos
```

```python
grupo=consecutivos.groupby('product_id')['groups'].count()
grupo.head(5)
```
